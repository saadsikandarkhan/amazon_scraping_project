# Use Case 3: Webscraping Function to scrape top 50 items in all departments categories and put it in a dataframe

```python
%%time

master_df = []
for item in bestseller_links_series:
    df = parsing_to_df_function(item)
    master_df.append(df)

master_df = pd.concat(master_df)
```
